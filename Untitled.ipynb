```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# Drop unnecessary columns
df = df.drop(['show_id' ,'title', 'director','description'], axis = 1)
```

```python
df.head()
```

```python
# Drop rows where important columns are missing
df = df.dropna(subset = ['date_added','country','rating', 'duration'])
```

```python
df.shape
```

```python
8807-7961
```

```python
# convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
# Remove leading/trailing spaces
df['date_added'] = df['date_added'].str.strip()
```

```python
# Convert to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['cast'] = df['cast'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.head(15)
```

```python
df.info()
```

```python
df.shape
```

```python
# Fix duration (numeric part only)
df['duration_int'] = df['duration'].str.extract('(\d+)').astype(float)
```

```python
df.head()
```

```python
df.to_csv('cleaned_netflix.csv', index=False)
```

```python
sns.set_style("whitegrid")	#Makes the background clean & grid visible
plt.rcParams['figure.figsize'] = (12,6)	#Sets all plots to a good readable size
```

## Movies VS TV Shows

```python
sns.countplot(x = 'type', data = df)
plt.title("Count of Movies vs TV shows on netflix", fontweight = "bold")
plt.xlabel("Type", fontweight = "bold")
plt.ylabel("Count", fontweight = "bold")
plt.savefig("movie_vs_tv.png",dpi = 300)
plt.show()
```

## Content Added Each Year (Trend)

```python
df['added_year'] = df['date_added'].dt.year

df['added_year'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added to Netflix Over the Years", fontweight = "bold")
plt.xlabel("Year Added",fontweight = "bold")
plt.ylabel("Count",fontweight = "bold")
plt.savefig("Content_Added_per_year.png", dpi = 300)
plt.show()
```

## Top 10 Genres

```python
genre_explode = df['listed_in'].str.split(', ').explode()

top_genres = genre_explode.value_counts().head(10)

sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres on Netflix", fontweight = "bold")
plt.xlabel("Count", fontweight = "bold")
plt.ylabel("Genre", fontweight = "bold")
plt.savefig("Top_10_Genres.png", dpi = 300)
plt.show()
```

## Top 10 Countries

```python
top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries Producing Netflix Content", fontweight = "bold")
plt.xlabel("Count", fontweight = "bold")
plt.ylabel("Country", fontweight = "bold")
plt.savefig("Top_10_Countries.png", dpi = 300)
plt.show()
```

##  Rating Distribution (PG-13, TV-MA, TV-14, etc.)


```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Content Rating Distribution", fontweight = "bold")
plt.xlabel("Count", fontweight = "bold")
plt.ylabel("Rating", fontweight = "bold")
plt.savefig("Rating_Distribution.png", dpi = 300)
plt.show()
```

## Movie Duration Distribution

```python
movies = df[df['type'] == 'Movie']

sns.histplot(movies['duration_int'], bins=30,kde = True)
plt.title("Distribution of Movie Durations (Minutes)", fontweight = "bold")
plt.xlabel("Minutes", fontweight = "bold")
plt.ylabel("Frequency", fontweight = "bold")
plt.savefig("Movie_Duration_Distribution.png", dpi = 300)
plt.show()
```

## TV Show Seasons Distribution

```python
tv = df[df['type'] == 'TV Show']

sns.countplot(x='duration_int', data=tv)
plt.title("Distribution of TV Show Seasons", fontweight = "bold")
plt.xlabel("Seasons", fontweight = "bold")
plt.ylabel("Count", fontweight = "bold")
plt.show()
```

## Compare Duration: Movies vs TV Shows

```python
sns.boxplot(x='type', y='duration_int', data=df)
plt.title("Movie Runtime vs TV Seasons", fontweight = "bold")
plt.xlabel("Type", fontweight = "bold")
plt.ylabel("Duration (Minutes or Seasons)", fontweight = "bold")
plt.show()
```
